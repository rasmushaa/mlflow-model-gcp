```python
model = mlflow.pyfunc.load_model("models:/BankingModel-main/1")

inputs = model.metadata.load_input_example()
display(inputs)
```

```python
registered_model_info = {}

client = mlflow.MlflowClient()

# Search for model versions by run_id
for rm in client.search_registered_models():
    for mv in client.search_model_versions(f"name='{rm.name}'"):
        if mv.run_id == model.metadata.run_id:
            print(f"\nModel Name: {mv.name}")
            print(f"Version: {mv.version}")
            print(f"Aliases: {mv.aliases}")
            print(f"Status: {mv.status}")
            print(f"Current Stage: {mv.current_stage}")
            break
```

```python
client = mlflow.MlflowClient()
registered_models = []

for model_vesrsion in client.search_model_versions(f"name='BankingModel-main'"):
    registered_models.append({
        "version": model_vesrsion.version,
        "run_id": model_vesrsion.run_id,
        "aliases": model_vesrsion.aliases,
        "tags": model_vesrsion.tags,
    })

registered_models
```
